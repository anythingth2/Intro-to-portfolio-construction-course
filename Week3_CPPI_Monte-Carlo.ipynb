# Portfolio Insurance (CPPI) and Drawdown Constriant

```python
ind_return_df = erk.get_ind_returns()
tmi_returns = erk.get_total_market_index_returns()
```

```python
risky_return_df = ind_return_df.loc['2000':, ['Steel', 'Fin', 'Beer']]
risky_return_df
```

```python
safe_return_df = pd.DataFrame().reindex_like(risky_return_df)
safe_return_df
```

```python
safe_return_df[:] = 0.03 / 12

start = 1000
floor = 0.8
```

```python
def run_cppi(risky_return_df, safe_return_df=None, m=3, start=1000, floor=0.8, riskfree_rate=0.03, max_drawdown=None):
    dates = risky_return_df.index
    n_steps = len(dates)
    account_value = start
    floor_value = start * floor

    if isinstance(risky_return_df, pd.Series):
        risky_return_df = pd.DataFrame(risky_return_df, columns=['R'])

    if safe_return_df is None:
        safe_return_df = pd.DataFrame().reindex_like(risky_return_df)
        safe_return_df.values[:] = riskfree_rate / 12

    account_history = pd.DataFrame().reindex_like(risky_return_df)
    cushion_history = pd.DataFrame().reindex_like(risky_return_df)
    risky_weight_history = pd.DataFrame().reindex_like(risky_return_df)

    peak = start
    for date, risky_returns in risky_return_df.iterrows():
        if max_drawdown is not None:
            peak = np.maximum(peak, account_value)
            floor_value = peak * (1 - max_drawdown)

        cushion = (account_value - floor_value) / account_value
        risky_weight = m * cushion
        
        risky_weight = np.maximum(np.minimum(risky_weight, 1), 0)
        safe_weight = 1 - risky_weight

        risky_allocation = account_value * risky_weight
        safe_allocation = account_value * safe_weight
        
        # Update the account value
        account_value = (risky_allocation * (1 + risky_return_df.loc[date])) + (safe_allocation * (1 + safe_return_df.loc[date]))
        cushion_history.loc[date] = cushion
        risky_weight_history.loc[date] = risky_weight
        account_history.loc[date] = account_value

    risky_wealth = start * (1 + risky_return_df).cumprod()
    backtest_result = {
        'wealth': account_history,
        'risky_wealth': risky_wealth,
        'risk_budget': cushion_history,
        'risky_allocation': risky_weight_history,
        'm': m,
        'start': start,
        'floor': floor,
        'risky_return': risky_return_df,
        'safe_return': safe_return_df
    }
    return backtest_result
```

```python
start = 1000
floor = 0.8
floor_value = start * floor

backtest_result = run_cppi(risky_return_df, m=3, start=start, floor=floor)

account_history = backtest_result['wealth']
cushion_history = backtest_result['risk_budget']
risky_weight_history = backtest_result['risky_allocation']
```

# Random Walk Generation

$$ \frac{S_{t+dt}-S_t}{S_t} = \mu dt + \sigma\sqrt{dt}\xi_t $$


```python
def geometric_brownian_motion(n_year=10,
                                n_scenario=1000,
                                mu=0.07,
                                sigma=0.15,
                                step_per_year=12,
                                start=100.0):
    dt = 1 / step_per_year
    n_step = step_per_year * n_year
    
    returns = np.random.normal(loc=mu*dt, scale=sigma*np.sqrt(dt), size=(n_step, n_scenario))
    returns[0] = 1
    
    return_df = pd.DataFrame(returns, columns=list(map(lambda i: f'scenario_{i}', np.arange(n_scenario))))
    price_df = start * (1 + return_df).cumprod()
    return price_df
```

# Interactive Plotting and Monte Carlo Simulations of CPPI

```python
import ipywidgets as widgets
from IPython.display import display
```

```python
def show_geometric_brownian_motion(n_scenario, mu, sigma):
    start = 100.0
    random_price_df = geometric_brownian_motion(n_scenario=n_scenario, mu=mu, sigma=sigma, start=start)
    ax = random_price_df.plot(legend=False, color='indianred', alpha=0.5, linewidth=2, figsize=(16, 6))
    ax.axhline(y=start, ls=':', color='black')
    ax.plot(0, start, marker='o', color='darkred', alpha=0.2)
```

```python
show_geometric_brownian_motion(100, 0.7, 0.3)
```

```python
interactive_plot = widgets.interactive(show_geometric_brownian_motion,
                                        n_scenario=(1, 100, 5),
                                        mu=(-0.2, 0.2, 0.01),
                                        sigma=(0, 0.3, 0.01)
)
display(interactive_plot)
```

```python
def show_cppi(n_scenario, mu, sigma, riskfree_rate=0.03, floor=0.2, m=3, max_drawdown=None, y_max=100):
    start = 100.0
    random_price_df = geometric_brownian_motion(n_scenario=n_scenario, mu=mu, sigma=sigma, start=start)

    backtest_result = run_cppi(risky_return_df=random_price_df.pct_change().dropna(), 
                                riskfree_rate=riskfree_rate, 
                                m=m,
                                floor=floor,
                                max_drawdown=max_drawdown,
                                start=start
                            )
    
    wealth = backtest_result['wealth']

    terminal_wealth = wealth.iloc[-1]

    tw_mean = terminal_wealth.mean()
    tw_median = terminal_wealth.median()
    failure_mask = np.less(terminal_wealth, start*floor)
    n_failures = failure_mask.sum()
    p_fail = n_failures/n_scenario

    e_shortfall = np.dot(terminal_wealth-start*floor, failure_mask)/n_failures if n_failures > 0 else 0.0


    fig, (wealth_ax, hist_ax) = plt.subplots(nrows=1, ncols=2, sharey=True, gridspec_kw={'width_ratios':[3,2]}, figsize=(24, 9))
    plt.subplots_adjust(wspace=0.0)
    
    wealth.plot(legend=False, alpha=0.5, linewidth=2, figsize=(16, 6), ax=wealth_ax)
    wealth_ax.axhline(y=start, ls=':', color='black')
    wealth_ax.axhline(y=start*floor, ls='--', color='red')
    wealth_ax.set_ylim(top=wealth.values.max()*y_max/100)
    
    terminal_wealth.plot.hist(bins=50, ec='w', orientation='horizontal', ax=hist_ax)
    hist_ax.axhline(y=start, ls=':', color='black')
    hist_ax.axhline(y=start*floor, ls='--', color='red')
    hist_ax.axhline(y=tw_mean, ls=":", color="blue")
    hist_ax.axhline(y=tw_median, ls=":", color="purple")
    hist_ax.annotate(f"Mean: ${int(tw_mean)}", xy=(.7, .9),xycoords='axes fraction', fontsize=24)
    hist_ax.annotate(f"Median: ${int(tw_median)}", xy=(.7, .85),xycoords='axes fraction', fontsize=24)
    if (floor > 0.01):
        hist_ax.axhline(y=start*floor, ls="--", color="red", linewidth=3)
        hist_ax.annotate(f"Violations: {n_failures} ({p_fail*100:2.2f}%)\nE(shortfall)=${e_shortfall:2.2f}", xy=(.7, .7), xycoords='axes fraction', fontsize=24)

```

```python
interactive_plot = widgets.interactive(show_cppi,
                                        n_scenario=(1, 1000, 5),
                                        mu=(-0.2, 0.2, 0.01),
                                        sigma=(0, 0.3, 0.01),
                                        riskfree_rate=(0.0, 0.05, 0.01),
                                        floor=(0, 1, 0.1),
                                        m=(1, 5, 0.5),
                                        max_drawdown=(0.0, 1.0, 0.1),
                                        y_max=(0, 100, 1)
)
display(interactive_plot)
```
